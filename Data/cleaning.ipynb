```python
videogame_df['Year'] = pd.to_numeric(videogame_df['Year'], errors='coerce', downcast='integer')

print(videogame_df.dtypes)
```

```python
videogame_df['Team'] = videogame_df['Team'].str.replace(r'[\[\]\'"]', '', regex=True)
videogame_df['Genres'] = videogame_df['Genres'].str.replace(r'[\[\]\'"]', '', regex=True)
```
